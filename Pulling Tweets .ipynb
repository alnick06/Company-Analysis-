```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()
sentiment_summary = dict()
for x in canwords:
    sentences = nltk.tokenize.sent_tokenize(x)
    for y in sentences:
        sentiment_score = sid.polarity_scores(y)
        if sentiment_score["compound"] == 0.0:
            sentiment_score["neu"] += 1
        elif sentiment_score["compound"] > 0.0:
            sentiment_score["pos"] += 1
        else:
            sentiment_score["neg"] += 1

    
```
